###### `IMDB_Rating`

Passa-se agora ao tratamento da variável relativa as avaliações dadas pelos usuários cadastrados no IMDb e que será a variável alvo a ser prevista pelo modelo de *Machine Learning* a ser desenvolvido adiante.

Segundo informação no portal do [IMDb](https://help.imdb.com/article/imdb/track-movies-tv/ratings-faq/G67Y87TFYYP6TWAV), a fórmula ponderada para cálculo do Top 250 oferece "verdadeira estimativa Bayesiana":

$$
WR=\frac{v}{v+m}×R+\frac{m}{v+m}×C
$$

$WR$: média ponderada (*weighted rating*).

$v$: número de votos recebidos.

$m$: mínimo número de votos requeridos (25 mil).

$R$: média da nota de todos os usuários para o filme.

$C$: nota média em toda a base de dados (\~6.8).

[Informação](https://www.imdb.com/list/ls076459507/) no portal do IMDb oferece um guia para dar nota aos filmes, apresentando detalhamento para cada nota da lista a seguir:

Diante disso, mantendo-se o intuito da preservação máxima dos dados, evitando-se a exclusão de linhas da tabela, a fim de preservar a integralidade de outras variáveis, sobretudo em uma base de dados que não é tão grande quantitativamente e levando em conta que os dados virão a alimentar modelos de *machine learning*, que necessitam da maior quantidade (e qualidade, claro) de dados para melhores predições, opta-se, novamente, por preencher os dados faltantes com uma média das medidas de centralidade (média, mediana e moda), o que não deve alterar muito a distribuição dos dados, uma vez que já se encontram mais ou menos centrados naquela região.

```python
# Média das medidas de centralidade para a variável `Meta_score`:
media=(76 + 78 + 79)/3
media
```

```python
# Preenchendo entradas ausentes na variável `Meta_score`:
DF=DF.fillna({'Meta_score':media})
# Divisão das notas dos críticos por 10 para melhor visualização e comparação com as notas do público:
# DF['Meta_score']=DF['Meta_score']/10
# Entradas únicas para a variável `Meta_score`:
print('Entradas únicas para a variável `Meta_score`:\n{}'.format(DF['Meta_score'].unique( )))
# Confirmando se ainda há valores nulos na variável `Meta_score`:
print('\nPresença de entradas nulas na variável `Meta_score`: {}\n'.format(DF['Meta_score'].isnull( ).values.any( )))
# Sumário estatístico para a variável `Meta_score`:
print('mode/moda: {:.2f}'.format(mode(DF['Meta_score'])))
DF['Meta_score'].describe( )
```

Há, ainda, fator extra. Embora, supostamente, seja uma variável que se refere a números (valores monetários), como visto nas propriedades do `DataFrame`, logo após o carregamento dos dados, essa variável está armazenada como `object`, ou seja, texto (`str`). Assim, antes de quaisquer procedimentos, necessário converter essa variável para números, de forma a que os dados possam ser melhor trabalhados, como feito com a variável anterior, por exemplo.

```python
# Contagem de itens nulos para a variável `Gross`:
print('Entradas nulas para a variável `Gross`: {}'.format(DF['Gross'].isnull( ).sum( )))
# Amostra de valores da coluna `Groos`:
DF['Gross'].head(10)
```

Essa é a variável com mais valores nulos, representando quase 17% dos dados. Verificar-se-á uma amostra de entradas com valores faltantes.

```python
# Convertendo o tipo de dados das entradas da variável `Gross`:
# Retirando as vírgulas:
DF['Gross']=DF['Gross'].str.replace('[,]','', regex=True)
# Convertendo para `float`:
DF['Gross']=DF['Gross'].astype(float)
print('Entradas nulas para a variável `Gross`: {}'.format(DF['Gross'].isnull( ).sum( )))
DF['Gross'].describe( )
```

O método `describe( )` do Pandas retorna um sumário estátisco de variáveis numéricas desconsiderando entradas com valores ausentes, assim, é possível verificar os valores estatísticos para a variável `revenue` antes de qualquer alteração.

Conforme apresentado, há 169 valores ausentes para essa variável. As entradas podem ser excluídas, mas implicaria perder quase 17% da base de dados.
```python
# Sumário estatístico com preenchimento pela média:
DF['Gross'].fillna(DF['Gross'].median( ), inplace=False).describe( )
```

Utilizando-se a média como elemento substitutivo dos dados ausentes, notam-se leves alterações, mesmo assim, considerando que a base não é tão grande e os valores nulos representam quase 1/5 dos dados, poderia ser um comprometimento aceitável para não descartar dados que poderiam ser úteis em outras variáveis. ##### Tratamento de *Outliers*

A fim de melhor preparar os dados para uso em modelos de *Machine Learning*, aprofundar-se-á a análise dos "pontos fora da curva" (*outliers*) com vista a verificar a possibilidade de melhor refinamento dos dados para modelos de predição.

Utilizar-se-á os interquartis e o *InterQuartile Range* (IQR) e o limite padrão de 1,5 vezes de distância do centro.

```python
# @title Fig. 16 - *Outiliers* `IMDB_Rating`
fig, ax = plt.subplots(figsize=(10, 3)   , tight_layout=True)
DF['IMDB_Rating'].plot(kind = 'box', vert =False, color='#000000')
ax.set_title('Distribuição original IMDB_Rating', fontsize=15, fontweight='bold')
for spine in ['left','top','bottom','right']:ax.spines[spine].set_visible(False)
plt.gca ( ).axes.get_yaxis( ).set_visible           (False)
plt.tick_params(axis='both', which = 'both', bottom=False)
plt.savefig('Fig16')
plt.show( )
# `IMDB_Rating`
imdbQ1 =DF['IMDB_Rating'].quantile(.25)
imdbQ3 =DF['IMDB_Rating'].quantile(.75)
imdbIQR=     imdbQ3    -    imdbQ1
print('\n            IQR IMDb: {:,.2f}'         .format(imdbIQR))
superiorIMDb=imdbQ3 + 1.5 * imdbIQR
inferiorIMDb=imdbQ1 - 1.5 * imdbIQR
print('Limite Superior IMDb: {:,.2f}'         .format(superiorIMDb),
    '\nLimite Inferior IMDb: {:,.2f}'         .format(inferiorIMDb))
print('\nDataset com {} entradas,'             .format(DF.shape[0]),
      '\ndas quais {} estão acima de {:,.2f},'.format( len(DF[DF['IMDB_Rating'] > superiorIMDb]), superiorIMDb),
      '\nrepresentando {:,.2f}% do dataset.'  .format((len(DF[DF['IMDB_Rating'] > superiorIMDb])/DF.shape[0])*100))
```

```python
# @title Fig. 17 - *Outiliers* `Meta_score`
fig, ax = plt.subplots(figsize=(10, 3)   , tight_layout=True)
DF['Meta_score'].plot(kind='box'  , vert =False, color=  '#000000')
ax.set_title('Distribuição original Meta_score', fontsize=15, fontweight='bold')
for spine in ['left','top','bottom','right']:ax.spines[spine].set_visible(False)
plt.gca ( ).axes.get_yaxis( ).set_visible           (False)
plt.tick_params(axis='both', which = 'both', bottom=False)
plt.savefig('Fig17')
plt.show( )
# `Meta_score`
metaQ1 =DF['Meta_score'].quantile(.25)
metaQ3 =DF['Meta_score'].quantile(.75)
metaIQR=     metaQ3    -    metaQ1
print('\n            IQR Meta:  {:,.2f}'         .format(metaIQR))
superiorMeta=metaQ3 + 1.5 * metaIQR
inferiorMeta=metaQ1 - 1.5 * metaIQR
print('Limite Superior Meta: {:,.2f}'          .format(superiorMeta),
    '\nLimite Inferior Meta:  {:,.2f}'         .format(inferiorMeta))
print('\nDataset com {} entradas,'             .format(DF.shape[0]),
      '\ndas quais {} estão abaixo de {:,.2f},'.format( len(DF[DF['Meta_score'] < inferiorMeta]), inferiorMeta),
      '\nrepresentando {:,.2f}% do dataset.'   .format((len(DF[DF['Meta_score'] < inferiorMeta])/DF.shape[0])*100))
```

Como visto, os *outliers* mais significativos estão no Faturamento e nos Votos, de forma que, passar-se-á a limpeza dos dados para remoção de *outliers*, com a efetiva exclusão de entradas que estejam acima ou abaixo dos limites superiores ou inferiores, respectivamente.

```python
# Limpeza da Base de Dados para Exlusão de Outliers:
df=DF.copy( )
df.drop(df[df['IMDB_Rating'  ] > superiorIMDb ].index, axis=0, inplace=True)
df.drop(df[df['Meta_score'   ] < inferiorMeta ].index, axis=0, inplace=True)
df.drop(df[df['No_of_Votes'  ] > superiorVotes].index, axis=0, inplace=True)
df.drop(df[df['Gross'        ] > superiorGross].index, axis=0, inplace=True)
df.drop(df[df['Released_Year'] > superiorYear ].index, axis=0, inplace=True)
df.drop(df[df['Runtime'      ] < inferiorTime ].index, axis=0, inplace=True)
df.drop(df[df['Runtime'      ] > superiorTime ].index, axis=0, inplace=True)
print('Dados Originais: {}'.format(DF.shape),
    '\nDados    Limpos: {}'.format(df.shape),
    '\nPercentual de dados descartados: {:.2f}%'.format(((1-df.shape[0]/DF.shape[0])*100)))
```

Por outro lado, o faturamento apresenta correlação negativa com as avaliações das obras audiovisuais, tanto pelo público, quanto pela crítica. Seria um indicativo de talvez ser melhor não produzir um filme muito conceituado, caso o objetivo seja faturamento?

### *Machine Learning* (ML)

Após vasta conciência situacional sobre os dados, pode-se começar a sequencia de processos com objetivo de realizar predições utilizando modelos de *Machine Learning*. Mais especificamente, quer-se prever a nota (*rating*) do IMDb, ou seja, a nota dada pelos usuários cadastrados no portal IMDb, a partir da amostra de dados disponibilizada pelo cliente. Como se busca um valor contínuo, trata-se de um **problema de regressão**, a ser solucionado por meio de aprendizado supervisionado, uma vez que o modelo irá aprender a partir de exemplos da base de dados, cuja classificação para cada entrada é conhecida, para então poder aplicar o aprendizado para realizar previsões sobre novas entradas, novos filmes ou filmes que não fizeram parte da amostra recebida.

Um dos primeiros passos para começar as etapas de ML é dividir o dataset em dados de treino e de teste, deixando 20% dos dados para teste do modelo.

Antes, porém, é ncessário selecionar apenas variáveis numéricas.

```python
df.columns
```

```python
# @title Divisão entre Treino e Teste
# Matrix de Variáveis (apenas numéricas, exceto variável alvo):
originalX=df.drop(['IMDB_Rating','Series_Title','Certificate','Genre','Overview','Director','Star1','Star2','Star3','Star4'], axis=1)
# Vetor Alvo:
originalY=df      ['IMDB_Rating']
# Separação Treino & Teste:
trainX, testX, trainY, testY=train_test_split(originalX, originalY, test_size=.2, train_size=.8, shuffle=True, stratify=None, random_state=random_state)
X=trainX
Y=trainY
x= testX
y= testY
print('Dimensionalidade dos Dados:',
      '\ntrainX:'    , X.shape,
      '\n testX:'    , x.shape,
      '\ntrainY:'    , Y.shape,
      '\n testY:'    , y.shape)
```

Efetuada a separação dos dados entre treino e teste, é necessário padronizar os dados, a fim de uniformizá-los, colocando-os em uma mesma escala.

```python
# @title Modelo
# Padronização dos Dados:
sc=StandardScaler( ).fit(X)
ts=StandardScaler( ).fit(Y.values.reshape(-1, 1))
X =sc.transform  (X)
x =sc.transform  (x)
Y =ts.transform         (Y.values.reshape(-1, 1))
# Instanciando o Modelo:
print('Dimensões de Entrada:', X.shape[1:],'\n')
model=Sequential (name='model')
model.add(Input  (shape=(X.shape[1:])))
model.add(Dense  (128, activation='relu'))
model.add(Dense  ( 64, activation='relu'))
model.add(Dense  ( 32, activation='relu'))
model.add(Dense  (  1))
model.summary( )
```

```python
# Treinamento (fit) do Modelo:
model.compile(optimizer='adam', loss='mean_squared_error')
history=model.fit(X, Y, epochs=100, shuffle=True)
```

```python
print('Avaliação do Modelo: {:.4f}'.format(model.evaluate(x, y)))
```

```python
# @title Salvando o Modelo
pickle.dump(model, open('LH_CD_MODEL.pkl','wb'))
print('Modelo salvo.')
```
