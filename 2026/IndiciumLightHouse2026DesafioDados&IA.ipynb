```python
# @title Funções
#   Formatação para exibição numérica em eixos de gráficos
def format(x , pos):
    if     x >= 1e6: return f'{x*1e-6:.1f}M'
    elif   x >= 1e3: return f'{x*1e-3:.0f}K'
    else           : return f'{x     :.0f}'
# Months
meses_map={'Jan':'Jan','Feb':'Fev','Mar':'Mar','Apr':'Abr','May':'Maio','Jun':'Jun',
           'Jul':'Jul','Aug':'Ago','Sep':'Set','Oct':'Out','Nov': 'Nov','Dec':'Dez'}
def mes_pt(x, p):
    # Converter posição numérica do Matplotlib de volta para data
    data  =mdates.num2date(x)
    # Pegar o mês abreviado em inglês (padrão do sistema)
    mes_en=data.strftime('%b')
    # Traduzir usando o dicionário (se não achar, mantém o original)
    mes_pt=meses_map.get(mes_en, mes_en)
    # Retornar no formato Mes/Ano
    return f'{mes_pt}/{data.strftime('%Y')}'
# Câmbio
def Cambio(start, end):
    try:
        start =pd.to_datetime(start, dayfirst=False, format='mixed').strftime('%Y-%m-%d')
        end   =pd.to_datetime( end , dayfirst=False, format='mixed').strftime('%Y-%m-%d')
        ticker=yf.Ticker('BRL=X')
        df    =   ticker.history(start=start, end=end)
        if df.empty:print('Yahoo Finance não retornou dados; verificar datas.')
        df    = df[['Close']].reset_index( )
        df.columns=['data','taxa_cambio']
        df['data']=pd.to_datetime(df['data']).dt.date
        return df
    except Exception as e:print(f'Erro: {e}')
# Mean Absolute Error (Erro Médio Absoluto)
def MAE (true, pred):
    true ,  pred   =np.array(true), np.array(pred)
    absolute_errors=np.abs  (true     -      pred)
    mae            =np.mean ( absolute_errors    )
    return mae
# Mean Absolute Percentual Error (Erro Percentual Médio Absoluto)
def MAPE(true, pred):
    true ,  pred   =np.array(true), np.array(pred)
    mask =  true  !=  0 # Evita divisão por zero caso haja dias sem venda
    mape =  np.mean(np.abs((true[mask] - pred[mask]) / true[mask]))*100
    return mape
print('Funções carregadas na memória.')
```

Combinando todas as informações:

```python
# Juntando com os dados de câmbio
start =df['data'].min( )
end   =df['data'].max( )
print(start, end)
taxa  =Cambio(start,end)
df       ['data']=pd.to_datetime  (df  ['data'], format='mixed').dt.date
taxa     ['data']=pd.to_datetime  (taxa['data'], format='mixed').dt.date
cambio           =pd.merge(df, taxa, on='data' ,    how= 'left')
cambio[ 'taxa_cambio']=cambio[ 'taxa_cambio'].ffill( )
cambio[ 'taxa_cambio']=cambio[ 'taxa_cambio'].bfill( )
cambio['custo_brl'   ]=cambio['custo_usd'   ]*cambio['qtd']*cambio['taxa_cambio']
cambio
```

```python
# SQL
conn  =   sql.connect (':memory:')
USD   .to_sql         ( 'custos' , conn, if_exists='replace', index=False)
BRL   .to_sql         ( 'vendas' , conn, if_exists='replace', index=False)
taxa  .to_sql         ( 'cambio' , conn, if_exists='replace', index=False)
cursor=  conn.cursor  (          )
query ='''
-- Unificando custos, cambio e vendas
WITH Reais AS (
    SELECT
        v.id_product AS id_produto,
        v.total AS receita_venda,
        (c.usd_price * v.qtd * tx.taxa_cambio) AS custo_brl
    FROM vendas v
    JOIN custos c  ON v.id_product= c.product_id
    JOIN cambio tx ON v.sale_date =tx.data
    ),
-- Verificando prejuízos
Perdas AS (
    SELECT
        id_produto,
        receita_venda,
        CASE
            WHEN receita_venda < custo_brl
            THEN (custo_brl - receita_venda)
            ELSE 0
        END AS prejuizo
    FROM Reais
    )
-- Agregação por id_produto
SELECT
    id_produto,
    ROUND  (                 SUM (receita_venda) , 2) AS  receita_total,
    ROUND  (SUM (prejuizo)                       , 2) AS prejuizo_total,
    ROUND ((SUM (prejuizo) / SUM (receita_venda)), 4) AS    perda_percentual
FROM Perdas
GROUP BY id_produto
ORDER BY perda_percentual DESC LIMIT 10
       '''
cursor        .execute (    query )
results=cursor.fetchall(          )
for result in results:print(result)
```

```python
report= pd.read_sql(query, conn)
report[   'receita_total']=report[   'receita_total'].map('{:,.2f}'.format)
report[  'prejuizo_total']=report[  'prejuizo_total'].map('{:,.2f}'.format)
report['perda_percentual']=report['perda_percentual'].map('{:.2%}' .format)
conn          .close   (          )
report
```

Inicialmente, buscar-se-á o Faturamento Total por Cliente, somando-se a coluna `total` para cada cliente ou `id_client`, no caso.

```python
# Faturamento Total (Revenue Ranking)
BRL.groupby('id_client')['total'].sum( ).sort_values(ascending=False).to_frame( ).rename(columns={'total':'faturamento_total'})\
                                                     .style.format('{:,.2f}').background_gradient(subset=['faturamento_total'], cmap='YlGn')
```

Assim, tem-se o ranking total de vendas por cliente, na qual o cliente `42` lidera. Agora. verificar-se-á a frequência – contagem total de transações – de cada cliente.

```python
# Frequência
BRL.groupby('id_client')['id'].count( ).sort_values(ascending=False).to_frame( ).rename(columns={'id':'frequencia'})\
                                                        .style.background_gradient(subset=['frequencia'], cmap='YlGn')
```

Observa-se que o cliente que mais realiza compras é o cliente `15`. Com esses dados, é possível calcular o Ticket Médio por cliente, dividindo-se o faturamento total pela frequência.

```python
# Agregando soma total (faturamento) & contagem de transações (frequência)
ticket=BRL.groupby('id_client').agg(faturamento_total=('total','sum'), frequencia=('id','nunique'))\
                               .assign(ticket_medio=lambda x:x['faturamento_total']/x['frequencia'])        \
                               .sort_values(by='ticket_medio', ascending=False)
# Ranking
ticket.style.format({'faturamento_total':'{:,.2f}',
                     'ticket_medio'     :'{:,.2f}'})\
                     .background_gradient(subset=['ticket_medio'], cmap='YlGn')
```

Agora que se tem o Ticket Médio de cada cliente, observa-se que o cliente `47`, quinto no ranking de faturamento, é o cliente com maior ticket médio, ou seja, gasta mais toda vez que faz compras.

### Resposta Questão 6

Para atender a demanda do Sr. Almir, será necessário adicionar dimensão de datas à base de dados `vendas_2023_2024.csv` utilizando um calendário de datas, por meio da função `strftime` para separar os dias, os dias da semana, os meses e o ano. Os dias da semana retornam um número de 0 a 6, correspondendo à sequência de domingo a sábado.

```python
# @title 6.1 – SQL
# Calendário
conn = sql.connect(':memory:')
BRL.to_sql  ('vendas' , conn, index=False, if_exists='replace')
conn.execute('DROP TABLE IF EXISTS calendario')
query='''
WITH RECURSIVE calendario(data_venda)AS(SELECT MIN (sale_date)          FROM vendas UNION ALL
                                        SELECT date(data_venda,'+1 day')FROM calendario
                                        WHERE       data_venda < (SELECT MAX(sale_date)FROM vendas))
SELECT data_venda,
    strftime     ('%Y', data_venda)AS ano,
    strftime     ('%m', data_venda)AS mes,
    strftime     ('%d', data_venda)AS dia,
    CASE strftime('%w', data_venda)
        WHEN '0' THEN 'domingo'
        WHEN '1' THEN 'segunda'
        WHEN '2' THEN 'terça'
        WHEN '3' THEN 'quarta'
        WHEN '4' THEN 'quinta'
        WHEN '5' THEN 'sexta'
        WHEN '6' THEN 'sábado'
    END AS dia_semana
FROM calendario
      '''
conn  .execute (query)
cal=pd.read_sql(query, conn)
cal#.style.set_caption('Tabela Calendário')
```

Com a tabela de datas pronta, com 731 linhas, pois 2024 foi bissexto, pode-se uni-la (`LEFT JOIN`) com a tabela de vendas, usando `COALESCE` para garantir que os dias sem vendas apareçam como `0` em vez de `null`.

```python
# LEFT JOIN
cal.to_sql('calendario', conn, index=False, if_exists='replace')
query='''
SELECT c.data_venda, c.mes, c.dia_semana, COALESCE(SUM(v.total), 0)AS valor_venda
FROM calendario c LEFT JOIN vendas v ON c.data_venda=v.sale_date
GROUP BY c.data_venda ORDER BY c.data_venda
      '''
joined=pd.read_sql(query, conn)
print(f'Dados consolidados. Dias sem faturamento: {len(joined[joined["valor_venda"]==0])}')
joined#.head(10).style.format({'valor_venda':'{:,.2f}'})
```

Antes, no entanto, como visto na fase de Análise Exploratória dos Dados (EDA), há muitos dados "fora da curva" (*outliers*) podendo gerar distorções na previsão dos modelos.

Uma boa técnica de tratamento de *outliers* é o uso *InterQuartile Range (IQR)*, focado na massa central dos dados, ou seja os 50% do meio. Boa prática com essa técnica é considerar um limite de 1,5 o intervalo interquartil, tanto para cima, quanto para baixo, com o cuidado de, no caso em estudo, o limite inferior pode ser no máximo zero, uma vez que não existe venda negativa.

```python
# Tratamento de OutLiers
Q1 =train['total'].quantile(.25)
Q3 =train['total'].quantile(.75)
IQR=Q3-Q1
print(f'IQR:             {IQR:,.2f}')
upper=Q3 + 1.5       *    IQR
lower=Q1 - 1.5       *    IQR
if lower <  0: lower = 0 # não existe venda negativa
print(f'Limite Superior: {upper:,.2f}\nLimite Inferior:         {lower:,.2f}')
outliers_count=train[(train['total'] <  lower)|(train['total'] >  upper)].shape[0]
train_clean   =train[(train['total'] >= lower)&(train['total'] <= upper)].copy ( )
print(f'Outliers removidos:      {outliers_count}\n\
Treino original:       {len(train)}\nTreino limpo:          {len(train_clean)}\n\
Teste preservado com    {len(teste)} registros reais')
```

    -- Unificando custos, cambio e vendas
    WITH Reais AS (
        SELECT
            v.id_product AS id_produto,
            v.total AS receita_venda,
            (c.usd_price * v.qtd * tx.taxa_cambio) AS custo_brl
        FROM vendas v
        JOIN custos c  ON v.id_product= c.product_id
        JOIN cambio tx ON v.sale_date =tx.data
        ),
    -- Verificando prejuízos
    Perdas AS (
        SELECT
            id_produto,
            receita_venda,
            CASE
                WHEN receita_venda < custo_brl
                THEN (custo_brl - receita_venda)
                ELSE 0
            END AS prejuizo
        FROM Reais
        )
    -- Agregação por id_produto
    SELECT
        id_produto,
        ROUND  (                 SUM (receita_venda) , 2) AS  receita_total,
        ROUND  (SUM (prejuizo)                       , 2) AS prejuizo_total,
        ROUND ((SUM (prejuizo) / SUM (receita_venda)), 4) AS    perda_percentual
    FROM Perdas
    GROUP BY id_produto
    ORDER BY perda_percentual DESC
    LIMIT 10;

        data_venda,
        strftime     ('%Y', data_venda) AS ano,
        strftime     ('%m', data_venda) AS mes,
        strftime     ('%d', data_venda) AS dia,
        CASE strftime('%w', data_venda)
            WHEN '0' THEN 'domingo'
            WHEN '1' THEN 'segunda'
            WHEN '2' THEN 'terça'
            WHEN '3' THEN 'quarta'
            WHEN '4' THEN 'quinta'
            WHEN '5' THEN 'sexta'
            WHEN '6' THEN 'sábado'
        END AS dia_semana
    FROM calendario;

    -- LEFT JOIN
    SELECT
        c.data_venda, c.mes, c.dia_semana,
        COALESCE (SUM (v.total), 0) AS valor_venda
    FROM calendario c
    LEFT JOIN vendas v ON c.data_venda = v.sale_date
    GROUP BY c.data_venda
    ORDER BY c.data_venda;

    -- Vendas por Dia da Semana
    SELECT
        dia_semana,
        SUM (valor_venda) AS faturamento_total,

    FROM faturamento

Construção de baseline: modelo simples, de modo a servir de comparação a qualquer inteligência artificial mais complexa. A média móvel de 7 dias, no caso, usa a média do que foi vendido nos 7 dias anteriores para estimar o próximo dia. Para tanto, fez-se grupamento de todas as vendas na mesma data, bem como considerou-se os dias em que não houve vendas. Para evitar vazamento de dados (data leakage), teve-se o cuidado de limitar os dados até o dia anterior ao dia para o qual se quer prever. Deve-se entender que o modelo proposto possui limitações: a média móvel ignora tendências de crescimento rápido e alta sazonalidade (intermitência de demanda), estando sempre em atraso a eventos correntes, podendo demorar até 7 dias, no caso, para refletir um acontecimento. A fragilidade do modelo ficou visualmente evidente no gráfico comparativo entre a realidade e as previsões, confirmada pelas métricas de erro (MAE e MAPE), que não deixam dúvidas ao refletirem erros muito acima do minimamente aceitável, o próprio erro percentual deixa exposta a limitação do modelo proposto. Ademais, para produto que não tem saída diária, o modelo é completamente impreciso, não podendo prever com confiaça a demanda de venda com base em janelas curtas (apenas 7 dias).
